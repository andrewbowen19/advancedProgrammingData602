## Extra Credit 2
We can use our SAT dataset for this, as we have the score columns poorly named (should be snake case), and they have missing values labelled as "s", which should be nans. Let's fix those

```python
sat
```

```python
# coercing df columns to snake case
new_cols = [x.replace(" ", "_").replace(".", "").lower() for x in sat.columns ]

sat.columns = new_cols

sat.head()
```

Now that we have our column names normalized, let's fill our missing "s" values with zeros, then coerce those score columns to integer types

```python
# Replace s column vals with 0
sat = sat.replace('s', 0)

# coercing score columns to integer types
sat['num_of_sat_test_takers'] = sat['num_of_sat_test_takers'].astype(int)
sat['sat_critical_reading_avg_score	'] = sat['sat_critical_reading_avg_score'].astype(int)
sat['sat_math_avg_score'] = sat['sat_math_avg_score'].astype(int)
sat['sat_writing_avg_score'] = sat['sat_writing_avg_score'].astype(int)

sat
```
